# Project description

*The virtual telephony service CallMeMaybe is developing a new function that will give supervisors information on the least effective operators. An operator is considered ineffective if they have a large number of missed incoming calls (internal and external) and a long waiting time for incoming calls. Moreover, if an operator is supposed to make outgoing calls, a small number of them is also a sign of ineffectiveness.*

## Task clarification
### Goal:
* determine how can we distinguish ineffective operators and identify them

### Assumption: such operators
* have a large number of missed incoming call
* have a long waiting time for incoming calls
* have a small number of outgoing calls (if is supposed to make ones)

# EDA

I want to use this function to have a proper look at different distributions

```python
def study_distribution(D, column, name):
    print('Studying distribution of', name)
    print('Most frequent value:\n', D[column].mode())
    print('Mean value:\n', round(D[column].mean(),2))
    print('Medaian value: \n', round(D[column].median(),2))
    if 'tariff_plan' in D.columns:
        fig = px.histogram(D, column, color = 'tariff_plan', title='Histogram distribution of '+ name)
        fig.show('png')
    else:
        fig = px.histogram(D, column, title='Histogram distribution of '+ name)
        fig.show('png')

    fig = px.scatter(D, column, title='Scatterplot distribution of '+ name)
    fig.show('png')
    print('Percentile values max: ', np.percentile(D[column], [90, 95, 99]))
    print('Percentile values min: ', np.percentile(D[column], [1, 5, 10]))
```

# Statistical hypotheses

## waiting_time

* H0: "Waiting time of 'effective' and 'ineffective' operators doesn't have statistically significant difference'
* H1:"Waiting time of 'effective' and 'ineffective' operators has statistically significant difference'

```python
effective_time = data_c[data_c.type =='effective']['waiting_time']
ineffective_time = data_c[data_c.type =='ineffective']['waiting_time']

alpha = .05

result = st.mannwhitneyu(effective_time, ineffective_time)[1]
 
if result < alpha:
    print('H0 rejected, there is a statustically significant difference between wating time length for effective and ineffective operators')
else:
    print('Failed to reject H0. No statustically significant difference between wating time length for effective and ineffective operators found')
effective_time.mean() < ineffective_time.mean()
```

### Conclusion

We found that there is a statustically significant difference between wating time length for effective and ineffective operators and that effective operators do have shorter waiting time

## missed calls

* H0: "Numbers of missed calls of 'effective' and 'ineffective' operators don't have a statistically significant difference'
* H1: "Numbers of missed calls of 'effective' and 'ineffective' operators have a statistically significant difference'

```python
missed_calls_effective = data_c[(data_c.type == 'effective') & (data_c.is_missed_call==True)]['calls_count']
missed_calls_ineffective = data_c[(data_c.type == 'ineffective')& (data_c.is_missed_call==True)]['calls_count']

result = st.mannwhitneyu(missed_calls_effective, missed_calls_ineffective)[1]
 
if result < alpha:
    print('H0 rejected, there is a statistically significant difference between amount of missed calls for effective and ineffective operators')
else:
    print('Failed to reject H0. No statistically significant difference between amount of missed calls for effective and ineffective operators found')
    
missed_calls_effective.mean() < missed_calls_ineffective.mean()
```
