The A/B test is dedicated to __testing changes related to the introduction of an improved recommendation system__<br>
To do so:
* I need to see whether within 14 days of signing up, users will show better conversion into product page views (the product_page event), product card views (product_card) and purchases (purchase). 

#### The interface_eu_test:
* Was launched at 2020-12-07
* Stopped at 2021-01-01
* New users were accepted until 2020-12-23 ( according to the Technical description they were accepted until 2020-12-23, but the description is outdated)
* 15% of the new users from the EU region
* Expected number of test participants: 6000

# Statistical significance of the difference in revenue

Let's formulate the null hypothesis:<br>
* H0: There is no statistically significant difference between A and B samples
* H1: There is a statistically significant difference between A and B samples

```python
A_sample_c = user_events_interface_A_c.details
B_sample_c = user_events_interface_B_c.details
```

```python
alpha = .05
results = st.mannwhitneyu(B_sample_c, A_sample_c,True, 'less')[1]

if results < alpha:
    print("Null hypothesis rejected. B performse worse than A")
else:
    print("Null hypothesis can not be rejected. There is no statistically significant difference between A and B samples")
```

# Statistically significant difference in proportions

* H0 hypothesis: there is no statistically significant difference between proportions of A and B groups.
* H1 hypothesis: there is one.

```python
proportions = pd.DataFrame({'A': user_events_interface_A_c.groupby(
    'event_name')['user_id'].nunique()})

proportions['B']= user_events_interface_B_c.groupby('event_name')['user_id'].nunique()

proportions = proportions.sort_values(by='B', ascending=False)
proportions
```

### Regular alpha

```python
alpha = 0.05
trials1 = user_events_interface_A_c.user_id.nunique()
trials2 = user_events_interface_B_c.user_id.nunique()
for event in event_list:
    success1 = proportions.loc[event, 'A']
    success2 = proportions.loc[event, 'B']
    p1 = success1/trials1
    p2 = success2/trials2
    difference = p1 - p2
    p_combined = (success1 + success2) / (trials1 + trials2)
    z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials1 + 1/trials2))
    distr = st.norm(0, 1)
    p_value = (1 - distr.cdf(abs(z_value))) * 2
    print('Share of A group is {: .2%}, for event: {}'.format(p1, event))
    print('Share of B group is {: .2%}, for event: {}'.format(p2, event))
    print(f'Success for A group is {success1}, for event: {event} out of {trials1} trials')
    print(f'Success for B group is {success2}, for event: {event} out of {trials2} trials')

    print(f'p-value for {event}: ', p_value)
    if (p_value < alpha):
        print(f"Rejecting the null hypothesis on event {event}: there is a significant difference between the proportions")
    else:
        print(f"Failed to reject the null hypothesis on event {event}: there is no reason to consider the proportions different")
    print()
```

### Šidák level

```python
number_of_tests = 4 #we have 4 events and 2 group only
alpha = 0.05

Šidák = 1 - pow((1-alpha),number_of_tests)
```

```python
event_list = ['login', 'product_page', 'product_cart', 'purchase']

trials1 = user_events_interface_A_c.user_id.nunique()
trials2 = user_events_interface_B_c.user_id.nunique()
for event in event_list:
    success1 = proportions.loc[event, 'A']
    success2 = proportions.loc[event, 'B']
    p1 = success1/trials1
    p2 = success2/trials2
    difference = p1 - p2
    p_combined = (success1 + success2) / (trials1 + trials2)
    z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials1 + 1/trials2))
    distr = st.norm(0, 1)
    p_value = (1 - distr.cdf(abs(z_value))) * 2
    print('Share of A group is {: .2%}, for event: {}'.format(p1, event))
    print('Share of B group is {: .2%}, for event: {}'.format(p2, event))
    print(f'Success for A group is {success1}, for event: {event} out of {trials1} trials')
    print(f'Success for B group is {success2}, for event: {event} out of {trials2} trials')

    print(f'p-value for {event}: ', p_value)
    if (p_value < Šidák):
        print(f"Rejecting the null hypothesis on event {event}: there is a significant difference between the proportions")
    else:
        print(f"Failed to reject the null hypothesis on event {event}: there is no reason to consider the proportions different")
    print()
```

### Conclusion
