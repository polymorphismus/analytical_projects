# Step 1. Downloading the data and its preprocessing


_We will use 6 <a id='libraries'>libraries</a>:_ 
* pandas: for data processing
* numpy: for calculations like varience, std, etc.
* matplotlib, seaborn, plotly express: for data visualisation
* datetime: for working with data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from datetime import timedelta
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

We have 3 datasets:
* Server logs with data on Yandex.Afisha visits from June 2017 through May 2018
* Dump file with all orders for the period
* Marketing expenses statistics
and we need to study the data they contain.

```python
try:
    visits_log = pd.read_csv(
        'https://code.s3.yandex.net//datasets/visits_log_us.csv',
        parse_dates=['End Ts', 'Start Ts'], dtype={'Device': 'category', 'Source Id': 'category'})

    orders_log = pd.read_csv(
        'https://code.s3.yandex.net//datasets/orders_log_us.csv', parse_dates=['Buy Ts'])

    costs = pd.read_csv(
        'https://code.s3.yandex.net//datasets/costs_us.csv', parse_dates=['dt'])
    print('Data extracted succesfully')

except FileNotFoundError:
    print('Ooops, the dateset not found.')

except pd.errors.EmptyDataError:
    print('Ooops, the dataset is empty.')
```

```python
datasets = {'visits_log': visits_log, 'orders_log': orders_log, 'costs': costs}

for dataset in datasets:

    print('The ' + dataset + ' dataset:')
    print(datasets[dataset].info(memory_usage='deep'))
    print(datasets[dataset].describe())
    print(datasets[dataset].duplicated().sum())
    datasets[dataset].columns = datasets[dataset].columns.str.lower()

    print('\n\n')
```

#### downcasting

```python
visits_log['uid'] = pd.to_numeric(visits_log['uid'], downcast='unsigned')
orders_log['uid'] = pd.to_numeric(orders_log['uid'], downcast='unsigned')
costs['costs'] = pd.to_numeric(costs['costs'], downcast='float')
costs['source_id'] = pd.to_numeric(costs['source_id'], downcast='integer')
```

#### rename

```python
visits_log = visits_log.rename(
    columns={'end ts': 'end_ts', 'source id': 'source_id', 'start ts': 'start_ts'})
orders_log = orders_log.rename(columns={'buy ts': 'buy_ts'})
```

```python
# this is an error, let's get rid of it.
visits_log[visits_log.end_ts < visits_log.start_ts]
```

```python
visits_log = visits_log[visits_log.end_ts > visits_log.start_ts]
```

### How often do users come back?

```python
first_visit = visits_log.groupby('uid')['start_ts'].min()
first_visit.name = 'first_visit_date'
```

```python
visits_log_ = pd.merge(visits_log, first_visit, on='uid')
```

```python
visits_log_['first_visit_month'] = visits_log_['first_visit_date'].astype('datetime64[M]')
```

```python
visits_log_['cohort_lifetime'] = visits_log_['month'] - visits_log_['first_visit_month']
visits_log_['cohort_lifetime'] = visits_log_['cohort_lifetime'] / np.timedelta64(1,'M')
visits_log_['cohort_lifetime'] = visits_log_['cohort_lifetime'].round().astype(int)
```

```python
cohorts = visits_log_.groupby(['first_visit_month', 'cohort_lifetime'])['uid'].nunique().reset_index()
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_visit_month','uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'}) 
```

```python
cohorts = cohorts.merge(initial_users_count,on='first_visit_month') 
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users'] 
```

```python
retention_pivot = cohorts.pivot_table(
    index='first_visit_month', columns='cohort_lifetime', values='retention', aggfunc='sum', fill_value=np.nan)
```

```python
retention_pivot.index = retention_pivot.index.strftime('%Y-%m')
```

```python
retention_index = retention_pivot.index.to_list()

a = ['mean_value']

retention_index.extend(a)
```

```python
retention_pivot = retention_pivot.append(retention_pivot.mean(skipna=True), ignore_index=True)
```

```python
retention_pivot.index = retention_index
```

```python
plt.figure(figsize=(13, 9))
plt.title('Cohorts: User Retention')
sns.heatmap(retention_pivot, annot=True, fmt='.2%', linewidths=1, linecolor='gray', cmap='icefire'); 
plt.show();
```

We can see here, mean values per cohorts lifetime remain more or less the same, from 4.26% to 6.38%, so retention is stable

### How much money do they bring? (LTV)


```python
cohort_sizes = orders_log_.groupby('month_first').agg(
    {'uid': 'nunique'}).reset_index()
cohorts = orders_log_.groupby(['month_first', 'order_month']).agg({
    'revenue': 'sum'}).reset_index()
```

```python
margin_rate = 0.5
```

```python
report = pd.merge(cohort_sizes, cohorts, on='month_first')
```

```python
report['age'] = (report['order_month'] - report['month_first']) / np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
```

```python
report['gp'] = report['revenue'] * margin_rate
report['ltv'] = report['gp'] / report['uid']
```
