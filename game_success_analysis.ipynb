### Data preprocessing

#### Column names  replacement

```python
games.columns = games.columns.str.lower()
```

#### Total sum for every game 

```python
games['total_sales'] = games.na_sales + \
    games.eu_sales + games.other_sales + games.jp_sales
```

## Cleaning Data and Data types conversion 

### Processing the missing values

#### Name

```python
games[games.name.isna()]
```

_We can see here the only two games with no names have the same platform and year of release, but no critic_score	user_score	rating. So let's just get rid of it_

```python
games = games.dropna(subset=['name'])
```

#### Year

```python
print(f"The percent of missing values in the year column is just {round(games.year_of_release.isnull().sum() / games.shape[0] * 100,1)}%, so we can get rid of them without really spoiling the data")
```

```python
round(games.year_of_release.isnull().sum() / games.shape[0] * 100,1)
```

```python
games = games.dropna(subset=['year_of_release'])
```

#### Scores and Ratings

We shouldn't fill the rating nor scores, it will spoil data, this is something we can't predict. So we have to get rid of them to make the data clear

```python
games['user_score'] = games['user_score'].replace('tbd', np.nan)
```

_To convert data, we need to get fill the missing values first. Let's see whether there is any strong corellation between columns. The corellation is generally considered strong when it's value is larger than 0.7, so let's see whether we have such in this dataset if we get rid of the missing values and look at the "clear data" only_

```python
games_clean = games.dropna()
games_clean = games_clean[games_clean.user_score.astype('string') != 'tbd']
games_clean.user_score = games_clean.user_score.astype('float')
```

```python
games_corr = games_clean.corr()

plt.figure(figsize=(13,9))
sns.heatmap(games_corr, cmap="RdBu_r", annot=True);
```

_We found strong corellation between_ 
- eu_sales and na_sales 
- eu_sales and other_sales
- na_sales and other_sales     
_but this doesn't help us with the missing values, because the corellated data is complete_

Looks like we have to put 0 in place of scores and use this data for all the tasks when we don't need to count on scores.

```python
games[['critic_score', 'user_score']] = games[[
    'critic_score', 'user_score']].fillna(0)
```

### convertion

```python
games.rating = games.rating.fillna('unknown')
```

```python
games['name'] = games['name'].astype('string')

games[['platform', 'genre', 'rating']] = games[[
    'platform', 'genre', 'rating']].astype('category')
```

```python
sale_names = [col for col in games.columns if 'sales' in col]
```

```python
games[['user_score'] + sale_names] = games[['user_score'] + sale_names].astype('float32')
games[['year_of_release', 'critic_score']] = games[[
    'year_of_release', 'critic_score']].astype('int')
```

```python
games.info()
```

```python
games.drop(games[(games.platform=='DS') & (games.year_of_release==1985)].index, inplace=True) #outlier from DS with non-proven data
```

```python
games.duplicated().sum()
```

### How long does it generally take for new platforms to appear and old ones to fade?

```python
top_platform = games.pivot_table(index='year_of_release', values='platform',
                                 aggfunc=lambda x: x.mode().iloc[0])
platforms_in_years = games.pivot_table(index='year_of_release', values='platform',
                                       aggfunc=['count', lambda x: pd.unique(x).to_list()])
platforms_in_years = platforms_in_years.join(top_platform)
platforms_in_years.columns = ['amount', 'platforms', 'top']
platforms_in_years
```

### The existing now platforms

```python
platforms = list(games.platform.unique())

current_platforms = platforms_in_years.loc[2016, 'platforms']
list(current_platforms)
```

```python
faded_platforms = pd.Series(platforms)[~pd.Series(platforms).isin(pd.Series(current_platforms))]
for platform in faded_platforms:
    print(platform, end=", ")
```

```python
platform_lifetime = games.sort_values(by='year_of_release').pivot_table(
    index='platform', values='year_of_release', aggfunc=['first', 'last'])
platform_lifetime.columns = ['first', 'last']
platform_lifetime['lifetime'] = platform_lifetime['last'] - \
    platform_lifetime['first'] + 1
platform_lifetime = platform_lifetime[platform_lifetime.index.isin(
    faded_platforms)]
platform_lifetime
```

```python
winner_platforms = pd.Series(platforms_in_years.top.unique())
winner_platforms = list(
    winner_platforms[~winner_platforms.isin(pd.Series(current_platforms))])
winner_platforms
```

```python
def plot_region_sales(index_criterium):
    regions = ['na_sales', 'eu_sales', 'jp_sales']
    for region in regions:
        sales = relevant_slice.pivot_table(
            index=index_criterium, values=region, aggfunc=['sum'])
        sales['share'] = sales[('sum', region)] / \
            sales[('sum', region)].sum() * 100
        if index_criterium == 'platform':
            sales=sales[sales.index.isin(relevant_slice_platforms)].sort_values(by='share', ascending=False)[:5]
        else:
            sales = sales.sort_values(by='share', ascending=False)[:5]

        fig = go.Figure()
        fig = px.bar(x = sales.index, y=sales[('sum', region)], color=sales.index)
        fig.update_layout(title_text='Sales by '+ index_criterium + ' in 2013-2016, ' + str(region), 
                      yaxis_title_text='total sales',
                      xaxis_title_text='platforms',
                         )
        fig.show('png')
        print(sales.sort_values(by='share', ascending=False)[:5])
        print(f"Variation of {region} is ", round(sales.share.var(), 2))
        print('*******************************')
        print()
```

```python
def stacked_region_sales(category):
    region_sales = relevant_slice.groupby(category)[['na_sales', 'eu_sales', 'jp_sales']].sum()
    region_sales = region_sales.replace(0.0, np.nan).dropna(axis=0, how='all')
    region_sales = region_sales.sort_values(by=['na_sales','eu_sales','jp_sales'], ascending=False).T
    print(region_sales)
    fig = go.Figure() 
    fig = px.bar(region_sales)
    fig.update_layout(title_text='Platform share per region in 2013-2016', 
                  yaxis_title_text='platforms',
                  xaxis_title_text='regions')
    
    fig.show('png')
```

```python
def plot_sales_share(category):
    platform_sales = relevant_slice.groupby(category)['total_sales'].sum().sort_values(ascending=False)
    if category == 'platform':
        platform_sales = platform_sales[platform_sales.index.isin(relevant_slice_platforms)]
    fig = go.Figure()
    fig = px.bar(platform_sales, color=platform_sales)
    fig.update_layout(title_text='Sales share per '+category+' platform in 2013-2016', 
                  yaxis_title_text='Number of games',
                  xaxis_title_text=category
    )
    fig.show('png')
```

# Step 5
 ## Test the following hypotheses:


AVERAGE USER RATINGS OF THE XBOX ONE AND PC PLATFORMS ARE THE  <a id='hypotheses_1'>SAME.</a>


I formulated the null hypotheses that the user ratings of the Xbox One and PC are the same

```python
Xbox_One = relevant_slice[relevant_slice['platform'] == 'XOne']
PC = relevant_slice[relevant_slice['platform'] == 'PC']
```

```python
st.levene(Xbox_One.user_score, PC.user_score, center='median')
```

```python
alpha = 0.05

results = st.ttest_ind(Xbox_One.user_score, PC.user_score)

print('p-value: ', results.pvalue)

if (results.pvalue < alpha):
    print("We reject the null hypothesis")
else:
    print("We can't reject the null hypothesis")
```

We can't reject the null hypothesis, which mean that the p-value is to high to be a coincidence. It means that a general user don't really distinguish between Xbox One and PC platforms 

AVERAGE USER RATINGS FOR THE ACTION AND SPORTS GENRES ARE  <a id='hypotheses_2'>DIFFERENT.</a>


I formulated the null hypotheses that the user ratings for the Action and Sports are the same

```python
Action = relevant_slice[relevant_slice['genre'] == 'Action']
Sports = relevant_slice[relevant_slice['genre'] == 'Sports']
```

```python
st.levene(Action.user_score, Sports.user_score, center='median')
```

```python
alpha = 0.05

results = st.ttest_ind(Action.user_score, Sports.user_score)

print('p-value: ', results.pvalue)

if (results.pvalue < alpha):
    print("We reject the null hypothesis")
else:
    print("We can't reject the null hypothesis")
```

We can't reject the null hypothesis, which mean that the p-value is to high to be a coincidence. The Action score and the average Sports score are in general the same.
