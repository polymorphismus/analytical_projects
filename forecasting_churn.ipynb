## project description
The gym chain Model Fitness is developing a customer interaction strategy based on analytical data. It has digitized a number of its customer profiles in order to fight churn that is not that obvious: we can calculate churn based on people who get rid of their accounts or don't renew their contracts, but sometimes a client leaves "on tiptoes" and we have no official confirmation that he is not interested anymore. My task is to analyze the data and come up with a customer retention strategy.<br>
I need to:
* Learn to predict the probability of churn (for the upcoming month) for each customer
* Draw up typical user portraits: select the most outstanding groups and describe their main features
* Analyze the factors that impact churn most
* Draw basic conclusions and develop recommendations on how to improve customer service:
* Identify target groups
* Suggest measures to cut churn
* Describe any other patterns you see with respect to interaction with customers

# Step 3.  Model to predict user churn

## LogisticRegression

I decided to drop the correlated columns because this lead to a better performance of the models (checked). Also, I've decided to use threshold, because our set is not balanced.

```python
X = gym.drop(['churn', 'avg_class_frequency_total', 'month_to_end_contract'], axis = 1)
y = gym['churn']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

scaler = StandardScaler()
scaler.fit(X_train)


X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)

model = LogisticRegression()

model.fit(X_train_st, y_train)


predictions = model.predict(X_test_st)
probabilities = model.predict_proba(X_test_st)[:,1]

#without theshold
print('Logistic regression: no threshold')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, predictions)))
print('F1: {:.2f}'.format(f1_score(y_test, predictions)))
print('ROC_AUC: {:.2f}\n'.format(roc_auc_score(y_test, probabilities)))


#with threshold
threshold = 0.4
custom_predictions = [0 if i<threshold else 1 for i in probabilities]



print('Logistic regression: threshold')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, custom_predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, custom_predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, custom_predictions)))
print('F1: {:.2f}'.format(f1_score(y_test, custom_predictions)))
print('ROC_AUC: {:.2f}\n'.format(roc_auc_score(y_test, probabilities)))
```

## RandomForest

```python
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0)
rf_model.fit(X_train_st, y_train)

rf_predictions = rf_model.predict(X_test_st)
rf_probabilities = rf_model.predict_proba(X_test_st)[:,1]

#without threshold
print('Random Forest Classifier: no threshold')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, rf_predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, rf_predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, rf_predictions)))
print('F1: {:.2f}'.format(f1_score(y_test, rf_predictions)))
print('ROC_AUC: {:.2f}\n'.format(roc_auc_score(y_test, rf_probabilities)))

#with threshold
threshold = 0.4
custom_predictions = [0 if i<threshold else 1 for i in rf_probabilities]

print('Random Forest Classifier: threshold')
print('Accuracy: {:.2f}'.format(accuracy_score(y_test, custom_predictions)))
print('Precision: {:.2f}'.format(precision_score(y_test, custom_predictions)))
print('Recall: {:.2f}'.format(recall_score(y_test, custom_predictions)))
print('F1: {:.2f}'.format(f1_score(y_test, custom_predictions)))
print('ROC_AUC: {:.2f}\n'.format(roc_auc_score(y_test, probabilities)))
```

## Conclusion

After implementing both models, I can see that the results are very close, but the LogisticRegression performs slightly better with all the metrics except for roc_auc, which is or 1% higher or the same. Probably the LogisticRegression function performs slightly better due to due to overfitting, which I tried to avoid by dropping 1 of paired columns with correlation more than 0.8. Some ensemble techniques are tend to reduce the chance of overfitting and this may be the reason for it showing lower results

# Step 4. Create user clusters

## Standardize the data

```python
scaler = StandardScaler()
X_sc = scaler.fit_transform(gym.drop(['churn', 'avg_class_frequency_total', 'month_to_end_contract'], axis = 1))
```

## Matrix of distances and dendrogram

```python
linked = linkage(X_sc, method = 'ward')

with plt.style.context('classic'):
    plt.figure(figsize=(15, 10))  
    dendrogram(linked, orientation='top')
    plt.title('Hierarchical clustering for GYM')
    plt.show()
```

### Remark

We've received a dendrogram showing the hierarchical relationship between objects. It is useful for us because by drawing a horizontal line on the longest vertical distance we can define the number of clusters:  it is the number of lines crossed by the dash line. The best choice if we don't want to dive in details too much would be 5 clusters - for a line drawn somewhere on 50 of y-level: to separate the high and mighty violet cluster into 2. Even though we can see that they are most similar, as the height of the link that joins them together is the smallest.

## K-means algorithm and predict customer clusters. 

```python
km = KMeans(n_clusters = 5, random_state=0) 
labels = km.fit_predict(X_sc) 

gym['cluster_km'] = labels
```

## Mean feature values for clusters

```python
gym_g = gym.groupby('cluster_km').mean().round(3)
gym_g
```

### plotly bar

```python
gym_g['clusters'] = gym_g.index
gym_g = gym_g.reset_index(drop=True)
```

```python
for col in gym_g.columns:
    fig = px.bar(gym_g, x='clusters', y=col)
    fig.update_layout(title_text=col)    
    fig.show('png')
```

### Remark

We have cluster 0 and 3 with the highest churn rate:0.559 and 0.403 consecuently. We can see that their contract_periods, month_to_end_contract and lifetime are the smallest (~2-3 times from cluster 5), as well as group_visits, they are the youngest of age, spend less avg_additional_charges_total. They visit gym more rare then all the other groups. So, relatively low rates is what distinguish the clients who are more prone to leave, those coming rare, no-group, low-spending young people who try the gym for the smalest period possible.
